```python
# проверка, что все элементы - численного типа
print("Типы данных в DataFrame:")
print(df1.dtypes)

# построение гистограмм распределения
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df1['x'], bins=10, color='blue', edgecolor='black')
plt.title('Гистограмма распределения x')
plt.xlabel('x')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(df1['y'], bins=10, color='green', edgecolor='black')
plt.title('Гистограмма распределения y')
plt.xlabel('y')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()

# построение "ящика с усами" для изучения выбросов
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=df1['x'])
plt.title('Ящик с усами для x')
plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=df1['y'])
plt.title('Ящик с усами для y')
plt.show()

# построение диаграммы рассеяния
plt.figure(figsize=(12, 6))
plt.scatter(df1['x'], df1['y'], color='red', edgecolor='black')
plt.title('Диаграмма рассеяния между x и y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

Исходя из построенных графиков, между параметрами х и у можно наблюдать линейную зависимость, поэтому целесообразно использовать корреляцию Пирсона.

```python
# КРИТЕРИЙ 2

# расчет коэффициента Пирсона

# расчет t-критерия

# вывод о степени корреляции величин и статистической значимости их взаимосвязи 
```

```python
# расчет коэффициента Пирсона
correlation_coefficient, p_value = stats.pearsonr(df1['x'], df1['y'])
print(f"Коэффициент корреляции Пирсона: {correlation_coefficient}")
print(f"p-value: {p_value}")

# расчет t-критерия
# вывод о степени корреляции величин и статистической значимости их взаимосвязи
n = len(df1)
t_statistic = correlation_coefficient * ((n - 2) ** 0.5) / ((1 - correlation_coefficient ** 2) ** 0.5)
print(f"t-статистика: {t_statistic}")
alpha = 0.05
if p_value < alpha:
    print("Существует статистически значимая корреляция между x и y.")
else:
    print("Нет статистически значимой корреляции между x и y.")
if abs(correlation_coefficient) < 0.3:
    correlation_strength = "Слабая"
elif abs(correlation_coefficient) < 0.7:
    correlation_strength = "Умеренная"
else:
    correlation_strength = "Сильная"
print(f"Корреляция: {correlation_strength} ({correlation_coefficient})")
```

```python
# КРИТЕРИЙ 3

# проверка, что все элементы - численного типа

# построение гистограмм распределения

# построение "ящика с усами" для изучения выбросов

# построение диаграммы рассеяния

# вывод о целесообразности использования корреляции Спирмена с обоснованием на основе предшествующих графиков
```

```python
# проверка, что все элементы - численного типа
print("Типы данных в DataFrame:")
print(df2.dtypes)

# построение гистограмм распределения
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df2['x'], bins=10, color='blue', edgecolor='black')
plt.title('Гистограмма распределения x')
plt.xlabel('x')
plt.ylabel('Частота')

plt.subplot(1, 2, 2)
plt.hist(df2['y'], bins=10, color='green', edgecolor='black')
plt.title('Гистограмма распределения y')
plt.xlabel('y')
plt.ylabel('Частота')

plt.tight_layout()
plt.show()

# построение "ящика с усами" для изучения выбросов
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=df2['x'])
plt.title('Ящик с усами для x')
plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=df2['y'])
plt.title('Ящик с усами для y')
plt.show()

# построение диаграммы рассеяния
plt.figure(figsize=(12, 6))
plt.scatter(df2['x'], df2['y'], color='red', edgecolor='black')
plt.title('Диаграмма рассеяния между x и y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

На графиках видно, что переменные не распределены нормально, взаимосвязь нелинейная и монотонная. Также на графиках видно, что присутствуют выбросы. Соответственно целесообразно использовать корреляцию Спирмена.

```python
# КРИТЕРИЙ 4

# расчет коэффициента Спирмена

# расчет t-критерия

# вывод о степени корреляции величин и статистической значимости их взаимосвязи 
```

```python
# расчет коэффициента Спирмена
spearman_corr, p_value = stats.spearmanr(df2['x'], df2['y'])
print(f'Коэффициент корреляции Спирмена: {spearman_corr}')
print(f'p-value: {p_value}')

# расчет t-критерия
# вывод о степени корреляции величин и статистической значимости их взаимосвязи
n = len(df2)
t_statistic = spearman_corr * np.sqrt((len(df2['x']) - 2) / (1 - spearman_corr**2))
print(f"t-статистика: {t_statistic}")
alpha = 0.05
if p_value < alpha:
    print("Существует статистически значимая корреляция между x и y.")
else:
    print("Нет статистически значимой корреляции между x и y.")
if 0 < abs(spearman_corr) < 0.1:
    correlation_strength = "Практически отсутствует"
elif 0.1 < abs(spearman_corr) < 0.3:
    correlation_strength = "Слабая"
elif 0.3 < abs(spearman_corr) < 0.5:
    correlation_strength = "Умеренная"
elif 0.5 < abs(spearman_corr) < 0.7:
    correlation_strength = "Заметная"
elif 0.7 < abs(spearman_corr) < 0.9:
    correlation_strength = "Сильная"
elif 0.9 < abs(spearman_corr) < 0.99:
    correlation_strength = "Очень сильная"
else:
    correlation_strength = "Практически линейная зависимость"
print(f"По шкале Чеддока корреляция: {correlation_strength} ({spearman_corr})")
```

```python
# КРИТЕРИЙ 5

# построение диаграммы рассеяния

# вывод о целесообразности использования корреляции Пирсона

# расчет коэффициента Пирсона
```

```python
# построение диаграммы рассеяния
plt.figure(figsize=(12, 6))
plt.scatter(df3['x'], df3['y'], color='red', edgecolor='black')
plt.title('Диаграмма рассеяния между x и y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

# вывод о целесообразности использования корреляции Пирсона
# Коэффициент Пирсона рассчитывается для нормального распределения, проверим это, используя тест Шапиро-Уилка на нормальность данных 
stats_x, p_value_x = stats.shapiro(df3['x']) 
stats_y, p_value_y = stats.shapiro(df3['y']) 
print(f'Тест Шапиро-Уилка (для х): статистика = {stats_x}, p-value = {p_value_x}')
print(f'Тест Шапиро-Уилка (для y): статистика = {stats_y}, p-value = {p_value_y}')
# Принятие решения на основе p-значения 
alpha = 0.05
if p_value_x < alpha: 
    print("Выборка X не нормальна") 
else: 
    print("Нельзя отклонить нулевую гипотезу о нормальности для Х") 
if p_value_y < alpha: 
    print("Выборка Y не нормальна") 
else: print("Нельзя отклонить нулевую гипотезу о нормальности для Y")

# расчет коэффициента Пирсона
correlation_coefficient, p_value = stats.pearsonr(df3['x'], df3['y'])
print(f"Коэффициент корреляции Пирсона: {correlation_coefficient}")
print(f"p-value: {p_value}")
if p_value < alpha:
    print("Существует статистически значимая корреляция между x и y.")
else:
    print("Нет статистически значимой корреляции между x и y.")
if abs(correlation_coefficient) < 0.3:
    correlation_strength = "Слабая"
elif abs(correlation_coefficient) < 0.7:
    correlation_strength = "Умеренная"
else:
    correlation_strength = "Сильная"
print(f"Корреляция: {correlation_strength} ({correlation_coefficient})")
```

```python
# КРИТЕРИЙ 6

# расчет коэффициентов линейной регрессии

# построение диаграммы рассеяния прогнозных и реальных значений на одном графике
```

```python
# расчет коэффициентов линейной регрессии
from sklearn.linear_model import LinearRegression
res = stats.linregress(df3['x'], df3['y'])
print(f'Y = {round(res.intercept, 3)} + {round(res.slope, 3)}*X')
print(f"Коэффициент наклона (b1): {round(res.slope, 3)}")
print(f"Свободный член (b0): {round(res.intercept, 3)}")

# построение диаграммы рассеяния прогнозных и реальных значений на одном графике
y_pred = res.intercept + res.slope * df3['x']
fig, ax = plt.subplots(figsize=(12, 6))
plt.title('Сравнение реальных и прогнозных значений')
ax.scatter(df3['x'], df3['y'], label='Реальные значения', color='blue', edgecolor='black')
ax.scatter(df3['x'], y_pred, label='Прогнозные значения', color='orange', edgecolor='black')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# расчет средней ошибки аппроксимации
X = sm.add_constant(df3['x'])
model = sm.OLS(df3['y'], X).fit()
print(f"Средняя ошибка аппроксимации (MEA): {np.mean(abs((df3['y'] - model.predict(X))/df3['y']))*100}")
```

```python
# КРИТЕРИЙ 14

# расчет индекса множественной корреляции

# расчет F-критерия

# вывод о статистической значимости уравнения в целом
```

```python
# расчет индекса множественной корреляции
multiple_correlation_index = model.rsquared
det_coef = multiple_correlation_index ** 0.5
print(f'Индекс множественной корреляции: {multiple_correlation_index}')
print(f'Коэффициент детерминации: {det_coef}')

# расчет F-критерия
f_statistic = model.fvalue
print(f'\nF-критерий: {f_statistic}')
# Расчет критического значения F
n = len(df4)
f_crit = stats.f.ppf(0.95, 2, n-2-1) # 2 - количество переменных
# P-value для F-статистики
p_value = 1 - stats.f.cdf(f_statistic, 2, n-2-1)
print(f"F-критическое = {f_crit}")
print(f"p-value = {p_value}")
```

Коэффициент детерминации (0.9958) показывает, что модель объясняет 99.58% вариаций зависимой переменной y. Индекс множественной корреляции (0.9979) указывает на сильную связь между факторами и полученным результатом. Также мы наблюдаем высокое значение F-критерия, что подтверждает значимость уравнения в целом.
