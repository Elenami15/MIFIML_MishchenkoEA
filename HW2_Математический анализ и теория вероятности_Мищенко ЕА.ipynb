**Задача 1**. (11 баллов)

В известном магазине техники представлено очень много экземпляров различной бытовой техники двух видов: с возможностью подключения к системе "Умный дом" и без такой возможности. Причем, известно, что эти два вида техники представлены в пропорции 1 : 3 соответственно. Случайным образом выбрали 7 экземпляров техники для розыгрыша среди участниц конкурса "Мисс Рукоделье - 2024".  Какова вероятность, что среди них не более двух экземпляров с возможностью подключения к системе "Умный дом"?

Подробно распишите решение и обоснуйте выбор распределения для моделирования сюжета задачи. Ответ округлите до трёх знаков.


**Решение:** Так как есть два возможных исхода (с подключением к системе "Умный дом" и без такой возможности) и фиксированное количество испытаний (случайным образом выбрано 7 экземпляров техники), мы можем использовать биноминальное распределение для решения задачи.

```python
n = 7  # общее количество экземпляров техники
p = 1/(1+3)  # вероятность выбора с подключением к системе "Умный дом", так как техника в пропорции 1:3
q = 1 - p  # вероятность выбора без системы "Умный дом"

# для вычисления вероятности, что среди n испытаний будет k=2 успехов, воспользуемся формулой для биноминального распределения
# воспользуемся для упрощения вычислений специальными функциями, округлим ответ до трех знаков
result = round(scipy.stats.binom.pmf(n-2,n,q)+scipy.stats.binom.pmf(n-1,n,q)+scipy.stats.binom.pmf(n,n,q), 3) 
print("Вероятность того, что среди выбранных экземпляров не более двух с возможностью подключения к 'Умному дому':", result)
```

**Задача 2** (30 баллов)

Дан закон распределения количества публикаций в которой группе телеграм-каналов:

$$
\begin{array}{|c|c|c|c|c|c|c|}
\hline X \text { (число публикаций) } & 0 & 1 & 2 & 3 & 4 & 5 \\
\hline \mathrm{P} \text { (вероятность) } & ? & 0,3 & 0,3 & 0,2 & 0,1 & 0,1 \\
\hline
\end{array}
$$

Найдите пропущенную вероятность для X = 0, математическое ожидание и дисперсию. Решите задачу "вручную" и с помощью python.

**Решение:**

Пропущенная вероятность для Х=0 вычисляем следующим образом (так как общая вероятность равна 1):

P(X = 0) = 1 - (0,3 + 0,3 + 0,2 + 0,1 + 0,1) = 1 - 1 = 0

По формуле вычисляем математическое ожидание (сумма произведений числа публикаций на их вероятность):

Е(Х) = 0 * 0 + 1 * 0,3 + 2 * 0,3 + 3 * 0,2 + 4 * 0,1 + 5 * 0,1 = 2,4

Дисперсия по формуле вычисляется следующим образом: D(X) = E(X^2) - (E(X))^2, следовательно, нужно сначала вычислить E(X^2):

E(X^2) = 0^2 * 0 + 1^2 * 0,3 + 2^2 * 0,3 + 3^2 * 0,2 + 4^2 * 0,1 + 5^2 * 0,1 = 7,4

Отсюда находим дисперсию:

D(X) = 7,4 - 2,4^2 = 7,4 - 5,76 = 1,64

```python
x = [0, 1, 2, 3, 4, 5]
p_x = [0, 0.3, 0.3, 0.2, 0.1, 0.1]  

# Математическое ожидание
e_x = sum(x_i * p for x_i, p in zip(x, p_x))

# E(X^2)
e_x2 = sum(x_i**2 * p for x_i, p in zip(x, p_x))

# Дисперсия
d_x = e_x2 - e_x**2

print("Пропущенная вероятность P(X = 0) =", p_x[0])
print("Математическое ожидание E(X):", e_x)
print("Дисперсия D(X):", d_x)
```

**Задача 3** (25 баллов)


Компания приобрела 10 000 новых ноутбуков. Вероятность поломки одного ноутбука в течение месяца равна 0,0002. Предполагается, что поломки отдельных ноутбуков происходят независимо друг от друга.

**Вопрос 1**. Найдите вероятность того, что за месяц сломается ровно 1 ноутбук.
Решение объясните. (5 баллов)


**Вопрос 2**. Компания планирует ввести программу расширенной гарантии, по которой она будет бесплатно ремонтировать или заменять ноутбуки, сломавшиеся в течение месяца. Если общее количество поломок за месяц превысит 5 ноутбуков, компания понесет значительные дополнительные расходы.

а) Найдите вероятность того, что за месяц сломается более 5 ноутбуков.Решение объясните (10 баллов)

б) Если стоимость ремонта или замены одного ноутбука составляет 500 долларов, вычислите математическое ожидание общих расходов компании на поломки за один месяц. (10 баллов)

**Решение:** Так как по условию есть независимое событие (поломка ноутбука) в фиксированный момент времени (в течение месяца) с учетом среднего количества раз, когда событие происходит в заданный момент времени, мы можем использовать для решения задачи распределение Пуассона.

```python
n = 10000  # количество ноутбуков
p = 0.0002  # вероятность поломки одного ноутбука
lambda_parametr = n * p  # среднее количество поломок

# Вопрос 1 
# вероятность того, что за месяц сломается ровно 1 ноутбук - для расчета используем формулу вероятности для распределения Пуассона, что за период времени произойдет k событий
k = 1
p_k = (lambda_parametr ** k * math.exp(-lambda_parametr)) / math.factorial(k)
print("Вероятность того, что сломается ровно 1 ноутбук:", p_k)

# Вопрос 2a: Вероятность того, что за месяц сломается более 5 ноутбуков = 1 - вероятность того, что за месяц сломается 5 и менее ноутбуков
p_5max = sum((lambda_parametr ** k * math.exp(-lambda_parametr)) / math.factorial(k) for k in range(6))
p_5min = 1 - p_5max
print("Вероятность того, что сломается более 5 ноутбуков:", p_5min)

# Вопрос 2б
# для распределения Пуассона математическое ожидание равно параметру лямбда
price = 500  # Стоимость ремонта или замены одного ноутбука
month_cost = lambda_parametr * price
print("Математическое ожидание общих расходов на поломки за один месяц:", month_cost, "долларов")
```
