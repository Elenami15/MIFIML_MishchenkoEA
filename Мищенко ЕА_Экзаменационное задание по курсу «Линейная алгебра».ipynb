### **Задание 1**

Исследуйте систему векторов на линейную зависимость или независимость без использования Python (вручную):

$$
a1=(1;8;-1), a2=(-2;3;3), a3=(4;-11;9)
$$

Подробно распишите решение, а в итоге сделайте вывод про то, являются ли векторы зависимыми или независимыми.

**Балл: 2**

### **Решение:**

Если $\exists$ коэффициенты $c1 \neq 0$, $c2 \neq 0$, $c3 \neq 0$, при которых выполняется равенство: $c1a1+c2a2+c3a3=0$ то векторы $a1, a2, a3$ будут линейно зависимы. Если таких коэффициентов нет, то векторы будут линейно независимы.

В данное уравнение подставим соответствующие значения векторов $a1, a2, a3$, в результате чего получаем систему:

$$
\begin{aligned}
& \quad c_1 - 2c_2 + 4c_3 = 0, \\
& \quad 8c_1 + 3c_2 - 11c_3 = 0, \\
& \quad -c_1 + 3c_2 + 9c_3 = 0.
\end{aligned}
$$

В матричном виде система запишется так:

$$
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix}
\cdot 
\begin{bmatrix}
c_1 \\
c_2 \\
c_3
\end{bmatrix}
=
\begin{bmatrix}
0 \\
0 \\
0
\end{bmatrix}
$$

Нам нужно вычислить определитель матрицы, чтобы установить, является ли она вырожденной (если определитель равен нулю) или невырожденной (определитель не равен нулю). Ранг вырожденной матрицы меньше ее размера, что будет говорить о том, что вектора матрицы линейно зависимы. Ранг невырожденной матрицы равен числу строк матрицы (в нашем случае $R=3$, то есть максимальный), то это будет говорить о том, что вектора матрицы линейно независимы.

$$
\text{det}
\begin{bmatrix}
1 & -2 & 4 \\
8 & 3 & -11 \\
-1 & 3 & 9
\end{bmatrix} =
1 \cdot 1 \cdot
\begin{vmatrix}
3 & -11 \\
3 & 9
\end{vmatrix}
+
(-1) \cdot 2 \cdot
\begin{vmatrix}
8 & -11 \\
-1 & 9
\end{vmatrix}
+
4 \cdot 1 \cdot
\begin{vmatrix}
8 & 3 \\
-1 & 3
\end{vmatrix} =
$$
$$=1\cdot60+2\cdot61+4\cdot27=60+122+108=290$$

Получили, что $\text{det}$ $\neq 0$, соответственно ранг матрицы $R=3$, следовательно, существует только одно решение системы уравнений, такое что $c1=0, c2=0, c3=0$.

### **Ответ:**

Система векторов $a1=(1,8,-1)$, $a2=(-2,3,3)$, $a3=(4,-11,9)$ линейно независима.

$$

\begin{vmatrix}

$$
$$
=(2-\lambda) \cdot (4-\lambda)(4-\lambda)-1) =
$$
$$
=(2-\lambda) \cdot (\lambda^2-8\cdot\lambda+15)

Приравниваем получившееся уравнение к нулю:

```python
# корреляционная матрица
corr_matrix = data.corr()
# тепловая карта для наглядности
plt.figure(figsize=(9, 9))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='viridis') 
plt.title('Корреляционная матрица')
plt.show()
```

```python
# расчет ранга корреляционной матрицы
rank = np.linalg.matrix_rank(corr_matrix)
# расчет количества столбцов
num_col = data.select_dtypes(include=[np.number]).shape[1]
# вывод на экран значения ранга и количества столбцов
print(f'Ранг корреляционной матрицы = {rank}')
print(f'Количество столбцов = {num_col}')
```

```python
# линейная зависимость - если ранг матрицы меньше количества столбцов
if rank < num_col:
    print("Между столбцами есть линейная зависимость")
else:
    print("Между столбцами нет линейной зависимости")

# Столбцы с корреляцией > 0.9
high_corr = corr_matrix [(corr_matrix > 0.9) & (corr_matrix < 1)]
high_corr = high_corr.dropna(how='all').dropna(axis=1, how='all')
print("\nСводная таблица столбцов с высокой корреляцией (>0.9):")
display(high_corr)
```

### Вывод:
Исходя из полученных данных ранг корреляционной матрицы $= 12$, количество столбцов $= 13$, соответственно, между столбцами есть линейная зависимость. Также проанализировав корреляционную матрицу, мы видим, что есть признаки с высокой корреляцией $(>0.9)$, а именно:
- temp (температура в Цельсиях) и atemp (температура по ощущениям в Цельсиях) - корреляция составляет $0.991702$.

## Задание 10
Разделите данные на предикторы и целевую переменную. Обучите модель с помощью классической линейной регрессии. Линейную регрессию необходимо реализовать без готовых библиотек.

Оцените качество модели с помощью хотя бы двух метрик качества.

Распишите решение с помощью кода на языке Python и напишите текстом интерпретацию полученного вывода.

**Балл: 5**

```python
# разделяем данные на предикторы и целевую переменную
X = data.drop(columns=["cnt"])  # предикторы
Y = data["cnt"]  # целевая переменная

# Добавляем столбец единиц для свободного члена
X = np.column_stack([np.ones(len(X)), X])
X = np.array(X)
Y = np.array(Y)
# Коэффициент линейной регрессии
lin_reg_koeff = np.linalg.inv(X.T @ X) @ X.T @ Y
# Прогнозное значение 
Y_pred = X @ lin_reg_koeff
# Среднеквадратичная ошибка
mse = np.mean((Y - Y_pred) ** 2)
# Коэффициент детерминации
deviation_total = np.sum((Y - np.mean(Y)) ** 2) # общее отклонение фактических значений от их среднего
sum_sq_remaind = np.sum((Y - Y_pred) ** 2) # сумма квадратов остатков
r2 = 1 - (sum_sq_remaind / deviation_total)

print("Метрики качества модели:")
print(f"Среднеквадратичная ошибка MSE: {mse}")
print(f"Коэффициент детерминации R^2: {r2}")
```

### Вывод:
$MSE$ измеряет среднюю квадратичную разницу между фактическими значениями и предсказанными значениями модели. Чем меньше значение $MSE$, тем лучше модель предсказывает данные. В данном случае, значение $MSE=753878.6752929664$, это довольно высокий показатель, что может указывать на то, что модель имеет значительные ошибки в предсказаниях.
Коэффициент детерминации $R^2$ показывает, какую долю вариации зависимой переменной объясняет модель (от 0 до 1). В нашем случае $R^2=0.7988398641466421$, значение близко к $0.8$, что означает, что около $80$% вариаций в данных объясняется моделью. Это довольно хороший показатель, указывающий на то, что модель хорошо справляется с задачей предсказания.
